#### Get relevant part of JSON and transform it into a *pandas* dataframe

```python
# assign relevant part of JSON to venues
venues = results['response']['venues']

# tranform venues into a dataframe
dataframe = json_normalize(venues)
dataframe.head()
```

#### Define information of interest and filter dataframe

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```

### Use nearest neighbour algorithm to estimate travel distance

#### Import packages and give staring points

```python
import geopy.distance

# Start location is original centre point
start_loc_df = pd.DataFrame({"lat":[latitude], "lng":[longitude]})

# Append start location to list of locations
df_new = pd.concat([start_loc_df,dataframe_filtered], sort=True)

# Re-index
df_new = df_new.reset_index(drop=True)

# Initialize distance
df_new['distance'] = 0

# Initialize ordered table
df_ordered = df_new.iloc[0:0]

# Calculate length of dataframe
locations = len(df_new.index)

# Insert extra ID for later loop
df_new['New_ID'] = range(1, 1 + len(df_new))

# Tuple list for displaying sequence

df_new
```

#### Loop through each point, determining distance, then iteratively add points to travel to

```python
# Add temp dataframe for ordering
df_temp = df_new

# Begin with starting point in row 0
closest_loc = 0

for j in range(1, locations):
    
    # Starting location
    start_coord = (df_temp.loc[closest_loc,'lat'], df_temp.loc[closest_loc,'lng'])

    # Extract first point for orderd table and remove from original table
    df_ordered = df_ordered.append(df_temp.loc[closest_loc])
    df_temp = df_temp.drop([closest_loc])
    
    # Re-order df_temp
    df_temp = df_temp.reset_index(drop=True)

    # Loop through all points, excluding first row
    rem_locs = len(df_temp.index)
    for i in range(1, rem_locs):
        coord2 = (df_temp.loc[i,'lat'], df_temp.loc[i,'lng'])
        df_temp.loc[i, 'distance'] = geopy.distance.geodesic(start_coord, coord2).km

    # Progressively find next location to visit
    closest_loc = df_temp['distance'].idxmin()

# Add last remaining point and start point
df_ordered = df_ordered.append(df_temp.loc[closest_loc])
df_ordered = df_ordered.append(df_ordered.loc[0])

# Remove rows where address is missing
df_ordered = df_ordered[df_ordered['address'].notna()]

df_ordered
```

#### Add 8 hour trip limit

```python
# Start ordered set at 0
df_ordered = df_new.iloc[0:0]

# Add temp dataframe for ordering
df_temp = df_new
df_temp['trip'] = 1

# Travel speed (km/hr)
travel_speed = 20

# Stop time for delivery (hours)
wait_time = 0.5

# Initialise k trips
k = 0

#while len(df_temp.index) > 10:
for k in range(1, 10):
    if len(df_temp.index) < 2:
        break
    
    # Begin with starting point in row 0
    closest_loc = 0
    start_coord = (df_temp.loc[closest_loc,'lat'], df_temp.loc[closest_loc,'lng'])
    
    #Start with time 0 and k trips
    time = 0

    # Calculate length of dataframe
    locations = len(df_temp.index)
    
    for j in range(1, locations):
        
        # Starting location
        start_coord = (df_temp.loc[closest_loc,'lat'], df_temp.loc[closest_loc,'lng'])

        # Extract first point for orderd table and remove from original table
        df_ordered = df_ordered.append(df_temp.loc[closest_loc])
        df_temp = df_temp.drop([closest_loc])
      
        # Re-order df_temp and df_ordered and put trip number in
        df_ordered = df_ordered.reset_index(drop=True)
        df_ordered.at[df_ordered.index[-1], 'trip'] = k
        df_temp = df_temp.reset_index(drop=True)
        
        # Loop through all points, excluding first row
        rem_locs = len(df_temp.index)
        for i in range(1, rem_locs):
            coord2 = (df_temp.loc[i,'lat'], df_temp.loc[i,'lng'])
            df_temp.loc[i, 'distance'] = geopy.distance.geodesic(start_coord, coord2).km

        # Progressively find next location to visit
        closest_loc = df_temp['distance'].idxmin()

        # Total time used up
        time = time + wait_time + df_temp['distance'].min() / travel_speed
        
        # Stop searching after 8 hour time limit reached
        if time > 8:
            break      

    # Add start point
    df_ordered = df_ordered.append(df_ordered.loc[0])
    df_ordered.loc[0, 'trip'] = k
    
#Remove last return point and add last point
df_ordered = df_ordered.append(df_temp)
df_ordered.at[df_ordered.index[-1], 'trip'] = k - 1
time = time + wait_time + df_temp['distance'].min() / travel_speed

print(time)

df_ordered
```
